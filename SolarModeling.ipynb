```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Data analysis of plant 1

```python
powerPlantOne = pd.read_csv("SolarData/Plant_1_Generation_Data.csv")
powerPlantOne.head()
```

```python
powerPlantOne.describe()
```

```python
powerPlantOne.shape
```

```python
print(powerPlantOne.isnull().sum())
```

```python
#how many different source keys do we have? 

plantCount = len(powerPlantOne['SOURCE_KEY'].unique())

print("Number of unique source keys " + str(plantCount))
```

```python
#plot date vs daily yield 

plt.figure(figsize=(12, 8))
plt.plot(powerPlantOne['DATE_TIME'], powerPlantOne['DAILY_YIELD'])
plt.xlabel('Date and Time')
plt.ylabel('Daily Yield')
plt.title('Daily Yield of Power Plant One')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))
axes[0,0].hist(powerPlantOne['DC_POWER'])
axes[0,0].set_title('DC_POWER')

axes[0,1].hist(powerPlantOne['AC_POWER'])
axes[0,1].set_title('AC_POWER')

axes[1,0].hist(powerPlantOne['DAILY_YIELD'])
axes[1,0].set_title('DAILY YIELD')

axes[1,1].hist(powerPlantOne['TOTAL_YIELD'])
axes[1,1].set_title('Total Yield')
```

```python
powerPlantOne['DATE_TIME'] = pd.to_datetime(powerPlantOne['DATE_TIME'])

# Extract the date part (day) from the 'time' column
powerPlantOne['DATE_TIME'] = powerPlantOne['DATE_TIME'].dt.date

# Calculate the number of unique days
unique_days_count = powerPlantOne['DATE_TIME'].nunique()

print("Number of unique days:", unique_days_count)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))
axes[0,0].hist(powerPlantOne['DC_POWER'])
axes[0,0].set_title('DC_POWER')

axes[0,1].hist(powerPlantOne['AC_POWER'])
axes[0,1].set_title('AC_POWER')

axes[1,0].hist(powerPlantOne['DAILY_YIELD'])
axes[1,0].set_title('DAILY YIELD')

axes[1,1].hist(powerPlantOne['TOTAL_YIELD'])
axes[1,1].set_title('Total Yield')
```

# Weather data plant one

```python
weatherDataOne = pd.read_csv('SolarData/Plant_1_Weather_Sensor_Data.csv')
```

```python
weatherDataOne.head()
```

```python
weatherDataOne.describe()
```

```python
print(weatherDataOne.isnull().sum())
```

```python
weatherDataOne['DATE_TIME'] = pd.to_datetime(weatherDataOne['DATE_TIME'])

# Extract the date part (day) from the 'time' column
weatherDataOne['DATE_TIME'] = weatherDataOne['DATE_TIME'].dt.date

# Calculate the number of unique days
unique_days_count = weatherDataOne['DATE_TIME'].nunique()

print("Number of unique days:", unique_days_count)
```

```python
weatherDataOne.shape
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))
axes[0,0].hist(weatherDataOne['AMBIENT_TEMPERATURE'])
axes[0,0].set_title('AMBIENT_TEMPERATURE')

axes[0,1].hist(weatherDataOne['MODULE_TEMPERATURE'])
axes[0,1].set_title('MODULE_TEMPERATURE')

axes[1,0].hist(weatherDataOne['IRRADIATION'])
axes[1,0].set_title('IRRADIATION')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

boxprops = dict(linewidth=2, color='blue')
medianprops = dict(linestyle='-', linewidth=2.5, color='red')
whiskerprops = dict(linestyle='--', linewidth=2, color='green')

# Use boxplot function to create the boxplot for AC_Power in powerPlantOne
powerPlantOne.boxplot(column='AC_POWER', ax=ax, boxprops=boxprops, medianprops=medianprops, whiskerprops=whiskerprops, positions=[0])

# Use boxplot function again to add the AC_Power from powerPlantTwo to the same plot
powerPlantTwo.boxplot(column='AC_POWER', ax=ax, boxprops=boxprops, medianprops=medianprops, whiskerprops=whiskerprops, positions=[1])

ax.set_xticks([0, 1])
ax.set_xticklabels(['powerPlantOne', 'powerPlantTwo'])
ax.set_title('Boxplot of AC_Power in powerPlantOne and powerPlantTwo')
ax.set_ylabel('AC_Power')

plt.show()
```
